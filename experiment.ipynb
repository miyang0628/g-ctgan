```python
# =====================================================
# Cell 1 [markdown]
# =====================================================
"""
# G-CTGAN Experiment Notebook

**논문**: 불균형 데이터 분류분석을 위한 가우시안 혼합 모델과 CTGAN을 결합한 오버샘플링 기법  
**저널**: 한국정보통신학회논문지 Vol.29, No.1 (2026)  
**DOI**: https://doi.org/10.6109/jkiice.2026.29.1.399

실행 순서: Cell을 위에서부터 순서대로 실행 (Shift+Enter)
"""

# =====================================================
# Cell 2 [code] — 환경 설정 및 모듈 로드
# =====================================================
import sys, os, importlib
sys.path.insert(0, os.path.abspath("."))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from pathlib import Path
from IPython.display import display

matplotlib.rcParams["font.family"] = "DejaVu Sans"
matplotlib.rcParams["axes.unicode_minus"] = False
pd.set_option("display.float_format", "{:.4f}".format)
pd.set_option("display.max_columns", 20)

Path("results").mkdir(exist_ok=True)

# 숫자로 시작하는 파일명 → importlib으로 로드
data_loader  = importlib.import_module("00_data_loader")
preprocess   = importlib.import_module("01_preprocess")
oversample   = importlib.import_module("02_oversample")
ablation_gmm = importlib.import_module("03_ablation_gmm")
train_eval   = importlib.import_module("04_train_evaluate")
report       = importlib.import_module("05_report")

print("Python :", sys.version)
print("작업 디렉토리 :", os.getcwd())
print("모듈 로드 완료 ✓")

# =====================================================
# Cell 3 [markdown]
# =====================================================
"""
## Step 00 — 데이터 로드
"""

# =====================================================
# Cell 4 [code]
# =====================================================
datasets = data_loader.load_all()
summary  = data_loader.summarize(datasets)
display(summary)
summary.to_csv("results/00_dataset_summary.csv", index=False)

# =====================================================
# Cell 5 [markdown]
# =====================================================
"""
## Step 01 — 전처리 (train/test 분할 + 스케일링)

- 분할 방식: **랜덤 stratified split** (random_state=42 고정)
- 행(고객/체납자) 단위 분할, 중복 없음
- Scaler는 train에서만 fit → test에 transform (data leakage 차단)
"""

# =====================================================
# Cell 6 [code]
# =====================================================
processed = preprocess.preprocess_all(datasets)

rows = []
for name, p in processed.items():
    rows.append({
        "dataset":         name,
        "train_n":         len(p["y_train"]),
        "test_n":          len(p["y_test"]),
        "train_minority%": round(p["y_train"].mean() * 100, 2),
        "test_minority%":  round(p["y_test"].mean()  * 100, 2),
        "n_features":      p["X_train"].shape[1],
    })
display(pd.DataFrame(rows))

# =====================================================
# Cell 7 [markdown]
# =====================================================
"""
## Step 03 — Ablation Study: GMM 군집 수 k 선택

k=2~10 범위에서 BIC(Bayesian Information Criterion) 최솟값 기준으로
각 데이터셋의 최적 군집 수를 결정합니다.
"""

# =====================================================
# Cell 8 [code]
# =====================================================
df_k = ablation_gmm.run_ablation(processed)
display(df_k)

# BIC 곡선 인라인 표시
img = plt.imread("results/03_bic_curves.png")
plt.figure(figsize=(12, 8))
plt.imshow(img)
plt.axis("off")
plt.tight_layout()
plt.show()

optimal_k = dict(zip(df_k["dataset"], df_k["optimal_k"]))
print("최적 k :", optimal_k)

# =====================================================
# Cell 9 [markdown]
# =====================================================
"""
## Step 04 — 실험 실행

### 실험 조건
| 조건 | 설명 |
|---|---|
| **Native** | 기법 고유 비율 (원 논문 재현) |
| **Unified** | 모든 기법 소수 범주 50% 통일 (심사자 2 지적 대응) |

> ⚠️ CTGAN/G-CTGAN 포함 시 수 시간 소요됩니다.
> 빠른 테스트는 `QUICK_TEST = True`로 설정하세요.
"""

# =====================================================
# Cell 10 [code]
# =====================================================
# QUICK_TEST = True  → CTGAN 계열 제외, 약 10~30분
# QUICK_TEST = False → 전체 실험, 수 시간 소요
QUICK_TEST = False

if QUICK_TEST:
    train_eval.METHODS = ["none", "smote", "adasyn", "gsmote"]
    print("[빠른 테스트 모드] CTGAN 계열 제외")
else:
    train_eval.METHODS = ["none", "smote", "adasyn", "gsmote", "ctgan", "gctgan"]
    print("[전체 실험 모드]")

df_res, df_hp = train_eval.run_experiments(processed, optimal_k)
print(f"\n총 실험 수: {len(df_res)}")
display(df_res.head(12))

# =====================================================
# Cell 11 [markdown]
# =====================================================
"""
## Step 05 — 결과 시각화 및 논문용 Table 생성
"""

# =====================================================
# Cell 12 [code] — 논문용 Table (조건 A · B)
# =====================================================
df_all    = pd.read_csv("results/04_all_results.csv")
t_native  = report.make_table(df_all, "native")
t_unified = report.make_table(df_all, "unified")

t_native.to_csv("results/05_table_native.csv",   index=False)
t_unified.to_csv("results/05_table_unified.csv", index=False)

print("=== 조건 A: Native (원 논문 재현) ===")
display(t_native)
print("\n=== 조건 B: Unified 50% (심사자 2 대응) ===")
display(t_unified)

# =====================================================
# Cell 13 [code] — AUC 비교 막대그래프
# =====================================================
report.plot_auc_comparison(df_all)

img = plt.imread("results/05_auc_comparison.png")
plt.figure(figsize=(16, 10))
plt.imshow(img)
plt.axis("off")
plt.tight_layout()
plt.show()

# =====================================================
# Cell 14 [code] — 비율 민감도 분석 (native vs unified)
# =====================================================
report.plot_ratio_sensitivity(df_all)

img2 = plt.imread("results/05_ratio_sensitivity.png")
plt.figure(figsize=(14, 5))
plt.imshow(img2)
plt.axis("off")
plt.tight_layout()
plt.show()

# =====================================================
# Cell 15 [code] — G-CTGAN 순위 요약
# =====================================================
sub = df_all[
    (df_all["condition"] == "unified") &
    (df_all["classifier"].isin(["RF", "LGBM"]))
]
rank = (sub.groupby(["dataset", "method"])["AUC"]
           .mean()
           .reset_index()
           .sort_values(["dataset", "AUC"], ascending=[True, False]))
rank["rank"] = rank.groupby("dataset")["AUC"].rank(
    ascending=False).astype(int)

print("=== 데이터셋별 G-CTGAN 순위 (RF+LGBM 평균 AUC 기준, unified) ===")
display(rank)

# =====================================================
# Cell 16 [code] — 하이퍼파라미터 요약
# =====================================================
hp = pd.read_csv("results/04_best_hyperparams.csv")
hp_summary = (hp[hp["condition"] == "unified"]
                .groupby(["dataset", "classifier"])
                ["best_params"].first()
                .reset_index())

print("=== 최적 하이퍼파라미터 (GridSearchCV 결과, unified) ===")
display(hp_summary)
hp_summary.to_csv("results/05_hyperparam_summary.csv", index=False)
print("저장 완료: results/05_hyperparam_summary.csv")
```
